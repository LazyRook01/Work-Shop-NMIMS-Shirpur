````python
import numpy as np
import pandas as pd
import sklearn
````

````python
churn_df = pd.read_csv("telecom_churn_clean.csv", index_col=0)
````

# Exercise 2

1. Import `KNeighborsClassifier` from `sklearn.neighbors`
2. Instantiate a KNeighborsClassifier called knn with 6 neighbors
3. Fit the classifier to the data using `.fit()` method

````python
from sklearn.neighbors import KNeighborsClassifier
X = churn_df[["total_day_charge", "total_eve_charge"]].values
y  = churn_df["churn"].values
print(X.shape,  y.shape)
````

````python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X,y)
````
